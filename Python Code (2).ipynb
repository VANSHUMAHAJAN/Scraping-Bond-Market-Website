```python
import bs4

from bs4 import BeautifulSoup

import requests
```

```python
website='https://fred.stlouisfed.org/categories'
```

```python
my_list=[]
```

```python
class scraper:
    def __init__(self,link):
        self.link=link
        html_text=requests.get(link).text
        soup=bs4.BeautifulSoup(html_text,'html')
        self.soup=soup
    

    def name(self):
      ###FOR RUNNING ALL THE CATEGORIES, THEN THEIR SUB CATEGORIES, THEN THEIR SUB CATEGORIES
        for x in range(0,len(self.soup.find_all('div',class_="col-xs-12 panel-menu")[0].find_all('li'))):
            
                       print(self.soup.find_all('div',class_="col-xs-12 panel-menu")[0].find_all('li')[x].a.text)
                       name=self.soup.find_all('div',class_="col-xs-12 panel-menu")[0].find_all('li')[x].a.text
                       sub_website_half=self.soup.find_all('div',class_="col-xs-12 panel-menu")[0].find_all('a')[x]['href'].split('/')[-1]
                       sub_website_2=website+'/'+sub_website_half
                       
                       self.sub_website_2=sub_website_2
                       sub_link=scraper(sub_website_2)
                       try:
                          sub_link.name()
                       except:
                          sub_link.listing_names()
                            
      
    
    def listing_names(self):
        
        try:
            ###FOR MULTIPLE PAGES
            for x in range(0,int(self.soup.find_all('a',class_="pager-link-gtm")[-2].text.replace('[','').replace(']',''))):
                a=self.link.split('/')[-1]
                page_web=f'{self.link}?cid={a}&t=&et=&pageID={x+1}'
                list_of_names_page=scraper(page_web)
                print('-------------------------------------------------------------------------------')
                for x in range(0,len(self.soup.find_all('div',class_="col-xs-12 col-sm-10"))):
                    
                    last_name=self.soup.find('table',class_="table-responsive").find_all('div',class_="col-xs-12 col-sm-10")[x].a.text
                    print(last_name)
                    
                    my_list.append(last_name)
                    my_list.append(1)
                    sub_website_half_4=self.soup.find('table',class_="table-responsive").find_all('div',class_="col-xs-12 col-sm-10")[x].a['href']
                    sub_website=website.split('/cat')[0]+sub_website_half_4
                    a=scraper(sub_website_4)
                    a.picking_data()
                    
        except:
          ###FOR SINGLE PAGE
            for x in range(0,len(self.soup.find_all('div',class_="col-xs-12 col-sm-10"))):
                    
                    last_name=self.soup.find('table',class_="table-responsive").find_all('div',class_="col-xs-12 col-sm-10")[x].a.text
                    print(last_name)
                    my_list.append(last_name)
                    my_list.append(1)
                    sub_website_half_4=self.soup.find('table',class_="table-responsive").find_all('div',class_="col-xs-12 col-sm-10")[x].a['href']
                    sub_website=website.split('/cat')[0]+sub_website_half_4
                    a=scraper(sub_website)
                    a.picking_data()
            
            
        
        
        
    def picking_data(self):
      ###FOR EXTRACTING THE INFORMATION OUT OF THE PAGES LIKE UNITS,FREQUENCY.TAGS ETC
        #FINAL UNITS,FREQUENCY
                a=self.soup.find_all('div',class_="panel-body")[0].find_all('p',class_="col-xs-12 pull-left")[0].strong.text
                b=self.soup.find_all('div',class_="panel-body")[0].find_all('p',class_="col-xs-12 pull-left")[0].span.text.replace('\n','').replace('\xa0','').replace('  ','')
                x=self.soup.find_all('div',class_="panel-body")[0].find_all('p',class_="col-xs-12 pull-left")[1].strong.text
                y=self.soup.find_all('div',class_="panel-body")[0].find_all('p',class_="col-xs-12 pull-left")[1].span.text.replace('\n','').replace('\xa0','').replace('  ','')
                
                print(f'{a} {b}')
                print(f'{x} {y}')
                my_list.append(f'{a} {b}')
                my_list.append(f'{x} {y}')
                
                new_list=[]
                # FINAL TAGS
                for x in range(0,len(self.soup.find_all('div',class_="series-tag-cloud")[-1].find_all('a',class_="series-tag fg-tag-lnk-gtm"))):
                    tags=self.soup.find_all('div',class_="series-tag-cloud")[-1].find_all('a',class_="series-tag fg-tag-lnk-gtm")[x].text.replace('\n','').replace('  ','')
                    print(f'TAG:{tags}')
                    new_list.append(f'TAG:{tags}')

                my_list.append(new_list)
                    
                    
                
                #FINAL PRICE
                date=self.soup.find_all('div',class_="pull-left meta-col")[0].find('span',class_="series-meta-value").text
                price=self.soup.find_all('div',class_="pull-left meta-col")[0].find('span',class_="series-meta-observation-value").text
                print(f'{date} {price}')
                my_list.append(f'{date} {price}')
                
                #FINAL LAST UPDATED
                updated=self.soup.find_all('div',class_="pull-left meta-col")[0].find('span',class_="updated-text").text.replace('\xa0','').replace('  ','')
                print(updated)
                my_list.append(updated)
                my_list.append(2)
                
                
```

##CONVERTING LIST TO A DICTIONARY

```python
my_dict={}
for index in range(0,len(my_list)):
    
    if (my_list[index]==1) or (my_list[index]==2):
        continue
    elif my_list[index+1]==1:
        key=my_list[index]
        my_dict[key]=[]
        continue
        
    else:
        my_dict[key].append(my_list[index])
    
    
```
